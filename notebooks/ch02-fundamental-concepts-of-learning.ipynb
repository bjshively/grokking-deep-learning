- Machine learning is a process through which computers learn to do tasks through processing data sets, rather than being explicitly programmed with rules (e.g. procedural programming)
- Deep learning is a subset of machine learning
- Deep learning is inspired by how brains work -- networks of simple neurons that are wired together
- Artificial intelligence overlaps with ML and DL, but does not encompass them

- Supervised learning -- Take known (labeled) data, use it to make predictions on unlabeled data
- Unsupervised learning -- Take unknown data and put it into groups (clusters)

- Parametric learning -- Make predictions by adjusting a fixed number of knobs (weights)
- Non-parametric learning -- Counting-based methods, where number of parameters is based on the input data, rather than preset

Deep learning algorithms are parametric, so non-parametric methods aren't covered in this book.
